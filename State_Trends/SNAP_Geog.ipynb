## Executive Summary 

The Supplemental Nutrition Assistance Program [(SNAP)](https://www.ers.usda.gov/topics/food-nutrition-assistance/supplemental-nutrition-assistance-program-snap/key-statistics-and-research) is the largest federal nutrition assistance program in the United States, providing monthly support to millions of households. Because eligibility and enrollment respond to economic conditions, SNAP participation is often used as an indicator of economic distress and recovery. This project examines how SNAP participation varied across U.S. states during the COVID-19 pandemic, highlighting differences in both the magnitude and timing of economic impacts across regions. Using a state-level panel dataset of monthly and yearly SNAP participation, the analysis explores national trends, geographic disparities, and structural differences in how states experienced changes in enrollment. The project is organized into four sections: national SNAP trends over time, geographic variation across states, COVID-era shock decomposition, and a regression analysis of COVID-related changes in SNAP participation. Together, these sections aim to clarify how SNAP participation evolved during the COVID-19 period and what factors help explain observed geographic variation.

## Background

In the United States, 48 million people face food insecurity. This means that 1 in 7 households are food insecure, and nearly 14 million kids don't have consistent access to daily meals [(Feeding America)](https://www.feedingamerica.org/hunger-in-america). Hunger exists everywhere, even in high-income countries like the United States, and can be caused by a variety of factors. From where we live to the cost of living, hunger creates health challenges and barriers to the future that are hard to break and overcome. 

The Supplemental Nutrition Assistance Program (SNAP, formerly called Food Stamps) exists to support the millions of Americans struggling to afford food. The SNAP program has existed for 60 years, reaching vulnerable populations with federal funds that are state-administered and designed to be countercyclical [(Harvard)](https://www.hks.harvard.edu/faculty-research/policy-topics/social-policy/explainer-understanding-snap-program-and-what-cuts). As the economy worsens, during periods like the Great Recession or the COVID-19 pandemic, SNAP enrollment is meant to increase, mitigating the impact that economic downturn has on the economy and alleviating costs for families and consumers. More or less, it acts as an automatic stabilizer, expanding during downturns to support and stimulate the economy. In turn, evidence has shown that SNAP has had a positive impact on Americans. SNAP reduces food insecurity, helps local economies, and lifts families out of poverty. It is also highly correlated with indicators like employment status, income, household composition, health outcomes, and food security [(CBPP)](https://www.cbpp.org/research/food-assistance/snap-is-linked-with-improved-health-outcomes-and-lower-health-care-costs#:~:text=SNAP%20is%20associated%20with%20improved,to%20build%20upon%20its%20successes.). Given this, examining SNAP trends in the United States in relation to other indicators can illustrate the program's economic and societal impacts, as well as the contexts in which it benefits Americans.

During the COVID-19 pandemic in particular, increased benefits and policy flexibilities helped to support millions of households during a time of dire need. SNAP participation rose from 36.9 million in February 2020 to over 43 million by May 2020 due to pandemic-related economic disruptions [(USDA)](https://www.ers.usda.gov/amber-waves/2021/october/coronavirus-covid-19-pandemic-transformed-the-u-s-federal-food-and-nutrition-assistance-landscape). While recent literature has examined the impact of SNAP during COVID-19 on participants, households, and the economy, less research has been done to understand which areas of the U.S. saw the most variation in their SNAP particpation during this period, and why that may be. The USDA and other institutions have looked at which areas of the U.S. are most dependent on food assistance programs like SNAP, but examining this during the COVID-19 period in particular could provide deeper insight into states' SNAP dependence post-COVID, and identify which states should rethink their food assistance support for future global crises [(USDA)](https://www.ers.usda.gov/data-products/charts-of-note/chart-detail?chartId=105123). 

## National SNAP Trends

By visualizing national SNAP trends, we can observe how participation has evolved over time and how major economic shocks are reflected in program usage. ### Historical Expansion

SNAP has evolved from a modest anti-hunger program into a major national stabilizer responsive to economic downturns. Food stamps started as an idea in 1939 as a response to the Great Depression. The Food Stamp Act was signed into law in 1964 and became available nationwide in 1974. After some restructuring, and to fight the stigma associated with "food stamp" usage, the program was officially renamed to the Supplemental Nutrition Assistance Program (SNAP) in 2008 [(USDA)](https://www.fns.usda.gov/snap/history). Following the Great Recession in 2008, SNAP saw its largest participation rates, with elevated participation driven by economic recovery, and has remained higher than historical rates since. 

![](Figures/snap_participation_trend.png)

### Participation and Federal Spending 

Not only has participation in SNAP increased, but the program has become more fiscally significant. SNAP responds to economic crises, so when Americans face economic hardship, spending on SNAP rises to accommodate the increased need for food assistance. In recent crises, like the Great Recession and the COVID-19 Pandemic, we can see how crisis spending relates to historical spending levels. 

![](Figures/snap_pands_trend.png)

### SNAP and Unemployment during Recessions

As SNAP reacts in response to economic conditions, this further correlates the safety net program with economic indicators like unemployment. As previously mentioned, SNAP acts as a counter-cyclical program. This means that participation and spending automatically increase during economic downturns (recessions) and decrease during economic expansions, acting as an "automatic stabilizer" for the economy. SNAP participation consistently rises during labor market contractions, suggesting strong responsiveness to macroeconomic instability. The shaded regions in the following figure represent periods of recession in the U.S., where you can see unemployment and SNAP participation both spike.

## State-Level SNAP Trends

While national SNAP trends reveal a broad expansion of the program during COVID-19, state-level patterns vary in impact. This section examines how SNAP participation shifted geographically across time, states, and regions. To analyze how SNAP participation changed across U.S. states during the COVID-19 pandemic, this project separates state participation into four analytical periods for analysis. The intention behind this was to evaluate how each phase of the COVID-19 period impacted SNAP. For each period, state SNAP participation is calculated as the average monthly number of participants across included years.

**Period Key:**

- *Pre-COVID Baseline (2018–2019):* Establishes pre-pandemic participation norms
- *COVID Shock Period (2020–2021):* Captures the immediate economic disruption and emergency policy response
- *Recovery / Transition (2022–2023):* Measures stabilization and post-crisis adjustment
- *Post Period (2024–2026):* Evaluates long-term persistence of SNAP participation changes

```python
# load datasets from the USDA Data Cleaning File
panel_df = pd.read_csv("Data/Generated_CSVs/snap_panel_2018_2026.csv")
COVID_df = pd.read_csv("Data/Generated_CSVs/snap_covid_shock_states.csv") # this is a more specific COVID shock dataset

# Panel is our starter dataset, which we will edit to create our plots
# panel_df.head(10)
```

```python
# Clean the panel dataset to prepare for figures 

# Remove incomplete year
panel_df = panel_df[panel_df["year"] < 2026].copy()

# PERIOD CLASSIFICATION
def assign_period(year):
    if year in [2018, 2019]:
        return "Pre"
    elif year in [2020, 2021]:
        return "Shock"
    elif year in [2022, 2023]:
        return "Recovery"
    else:
        return "Post"

panel_df["period"] = panel_df["year"].apply(assign_period)

# REGIONAL MAPPING (USDA-style codes)
region_map = {
    "NERO": "Northeast",
    "SERO": "Southeast",
    "MWRO": "Midwest",
    "SWRO": "Southwest",
    "MPRO": "Mountain Plains",
    "WRO": "West"
}

panel_df["region_clean"] = panel_df["region"].map(region_map)

# STATE CODES (for choropleths)
state_map = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR",
    "California": "CA", "Colorado": "CO", "Connecticut": "CT",
    "Delaware": "DE", "District of Columbia": "DC", "Florida": "FL",
    "Georgia": "GA", "Hawaii": "HI", "Idaho": "ID", "Illinois": "IL",
    "Indiana": "IN", "Iowa": "IA", "Kansas": "KS", "Kentucky": "KY",
    "Louisiana": "LA", "Maine": "ME", "Maryland": "MD", "Massachusetts": "MA",
    "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS",
    "Missouri": "MO", "Montana": "MT", "Nebraska": "NE", "Nevada": "NV",
    "New Hampshire": "NH", "New Jersey": "NJ", "New Mexico": "NM",
    "New York": "NY", "North Carolina": "NC", "North Dakota": "ND",
    "Ohio": "OH", "Oklahoma": "OK", "Oregon": "OR", "Pennsylvania": "PA",
    "Rhode Island": "RI", "South Carolina": "SC", "South Dakota": "SD",
    "Tennessee": "TN", "Texas": "TX", "Utah": "UT", "Vermont": "VT",
    "Virginia": "VA", "Washington": "WA", "West Virginia": "WV",
    "Wisconsin": "WI", "Wyoming": "WY"
}

panel_df["state_code"] = panel_df["state"].map(state_map)

# YEARLY AGGREGATION
yearly_df = (
    panel_df.groupby(["state", "state_code", "year"])["persons"]
    .mean()
    .reset_index()
)

# PRE-COVID BASELINE
pre_avg = (
    yearly_df[yearly_df["year"].isin([2018, 2019])]
    .groupby("state")["persons"]
    .mean()
)

yearly_df["pre_avg"] = yearly_df["state"].map(pre_avg)

# PCT CHANGE FROM BASELINE
yearly_df["pct_change"] = (
    (yearly_df["persons"] - yearly_df["pre_avg"])
    / yearly_df["pre_avg"]
) * 100
```

## Regression Analysis

To complement the descriptive analysis, a simple OLS regression framework is used to quantify the relationship between SNAP participation, time trends, and the COVID-19 shock. While the previous sections highlight patterns in national and state-level SNAP usage, regression analysis allows us to test whether these patterns remain statistically significant once we control for baseline regional differences and underlying trends over time. In particular, the models examine whether COVID-19 represents a structural break in SNAP participation and whether its effects vary across U.S. regions. Together, these specifications provide a more rigorous test of the patterns observed in the data and help distinguish between general trends and shock-driven changes. 

**Below are three models:**

1. *Model 1: Baseline* - How does SNAP vary over time and across regions?
2. *Model 2: COVID* - What changed during COVID, *after* controlling for time + region?
3. *Model 3: Region* - Does COVID affect SNAP differently across regions?


```python
reg_df = panel_df.copy()

# remove incomplete year if needed
reg_df = reg_df[reg_df["year"] < 2026].copy()

# log transform (stabilizes scale differences across states)
reg_df["log_persons"] = np.log(reg_df["persons"])

# time trend (captures long-run growth/decline)
reg_df["time_trend"] = reg_df["year"] - 2018

# COVID indicator
reg_df["covid"] = (reg_df["year"] >= 2020).astype(int)
```

```python
# Baseline Model - Ignoring COVID for a second, how does SNAP vary over time and across regions?
model1 = smf.ols(
    "log_persons ~ time_trend + C(region)",
    data=reg_df
).fit()

print(model1.summary())
```

**Model 1 Analysis:** 

The regression dataset is constructed from the full panel of state-level SNAP participation data from 2018–2025, excluding 2026 due to incomplete observations. The model uses 2018 as the baseline year, which serves as the starting point for measuring time trends prior to the COVID-19 shock.

A time trend variable is defined as the number of years since 2018 to capture gradual changes in SNAP participation over time. To identify the structural impact of the pandemic, a COVID indicator is included, taking the value of 1 for years 2020 and later, and 0 for pre-pandemic years (2018–2019). This classification allows the model to isolate differences in SNAP participation associated with the onset of COVID-19 from underlying pre-existing trends.

Finally, the dependent variable is log-transformed SNAP participation to account for differences in scale across states and to improve interpretability in percentage terms.

When we look at the time_trend coefficient (0.0267), we see that SNAP participation is increasing over time, even after controlling for region differences, because this is a positive value. Additionally, there are substantial and statistically significant differences across U.S. regions, with the Southeast and Midwest exhibiting higher SNAP participation relative to the baseline region, and the Mountain Plains region showing substantially lower participation.

```python
# model 2 - What changed during COVID AFTER controlling for time + region?
model2 = smf.ols(
    "log_persons ~ time_trend + covid + C(region)",
    data=reg_df
).fit()

print(model2.summary())
```

**Model 2 Analysis:**

To quantify the impact of COVID-19 on SNAP participation, a log-linear OLS regression is estimated, controlling for time trends and regional fixed effects. The results indicate that COVID is associated with a statistically significant increase in SNAP participation of approximately 27% (p < 0.001), after accounting for baseline temporal and geographic variation. Once COVID is included, the previously observed time trend becomes statistically insignificant, suggesting that much of the apparent growth in SNAP participation is attributable to the pandemic shock rather than a steady pre-existing trend.

Here we see that the COVID coefficient = 0.2381 --> exp(0.2381) - 1 ≈ 0.27 -> which is approximately a 27% increase in SNAP participation. However, we also see that once COVID is included, there is no clear underlying time trend in SNAP participation because time_trend = -0.0095, but p = 0.348 (NOT significant). It is also important to note that regional variations still exist, but are not much different from the period model.
